```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'magic04.data'
df = pd.read_csv(file_path, header=None)
df.head(10)
```

```python
def maximum(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns maximum value
    #Do not print anything to stdout
    file_path = 'filename'
    return df[attribute_number].max(axis = 0)


def minimum(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns minimum value
    #Do not print anything to stdout
    file_path = 'filename'
    return df[attribute_number].min(axis = 0)
    
def mean(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns mean value
    #Do not print anything to stdout
    file_path = 'filename'
    return df[attribute_number].mean(axis = 0)
    
def std(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns standard deviation value
    #Do not print anything to stdout
    file_path = 'filename'
    return df[attribute_number].std(axis = 0)
    
def q1(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns Q1 value
    #Do not print anything to stdout
    file_path = 'filename'
    return np.percentile(df[attribute_number], 25) 
    
    
def q3(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns Q3 value
    #Do not print anything to stdout
    file_path = 'filename'
    return np.percentile(df[attribute_number], 75) 

    
def median(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns median value
    #Do not print anything to stdout
    file_path = 'filename'
    return df[attribute_number].median(axis = 0)
    
    
def iqr(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns IQR value
    #Do not print anything to stdout
    file_path = 'filename'
    return np.percentile(df[attribute_number], 75)-np.percentile(df[attribute_number], 25) 
    
    
def count(filename, attribute_number):
    #Input: Filename a string and the attribute number 1 to 10
    #Output: Returns count value
    #Do not print anything to stdout
    file_path = 'filename'
    return df[attribute_number].count()
    

```

```python
test_At=6
print("Max at",test_At+1,"th attributes",maximum('magic04.data',test_At))
print("Min at",test_At+1,"th attributes",minimum('magic04.data',test_At))
print("Mean at",test_At+1,"th attributes",mean('magic04.data',test_At))
print("STD at",test_At+1,"th attributes",std('magic04.data',test_At))
print("Q1 at",test_At+1,"th attributes",q1('magic04.data',test_At))
print("Q3 at",test_At+1,"th attributes",q3('magic04.data',test_At))
print("Median at",test_At+1,"th attributes",median('magic04.data',test_At))
print("IQR at",test_At+1,"th attributes",iqr('magic04.data',test_At))
print("Count at",test_At+1,"th attributes",count('magic04.data',test_At))


```
